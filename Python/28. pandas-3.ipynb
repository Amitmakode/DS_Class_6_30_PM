```python
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4,5],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', "abc"],
    'dept_id': [10, 20, 10, 30,40]
})

departments = pd.DataFrame({
    'dept_id': [10, 20, 30],
    'dept_name': ['Engineering', 'Marketing', 'Sales']
})
```

```python
employees
```

```python
departments
```

```python
inner_merge = pd.merge(employees, departments, on ="dept_id" )
inner_merge
```

```python
inner_merge = pd.merge(employees, departments, on ="dept_id", how="inner" )
inner_merge
```

```python
inner_merge = pd.merge(employees, departments, on ="dept_id", how="left" )
inner_merge
```

```python
inner_merge = pd.merge(employees, departments, on ="dept_id", how="right" )
inner_merge
```

```python
high_freq_dates = pd.date_range('2023-01-01', periods=1000, freq='H')
high_freq_data = pd.DataFrame({
    'value': np.random.randn(1000),
    'volume': np.random.randint(100, 1000, 1000)
}, index=high_freq_dates)
```

```python
high_freq_dates
```

```python
high_freq_data
```

```python
# Resample to lower frequency (downsampling)
daily_data = high_freq_data.resample('D').agg({
    'value': 'mean',      # Average value per day
    'volume': 'sum'       # Total volume per day
})
```
