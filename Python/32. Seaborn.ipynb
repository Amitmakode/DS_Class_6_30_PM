```python
import seaborn as sns       # Seaborn for advanced statistical visualization
import matplotlib.pyplot as plt  # Matplotlib for fine-tuned figure/customization
import numpy as np          # Numpy for data creation/manipulation
import pandas as pd         # Pandas for DataFrame support (Seaborn works easily with them)
```

```python
# ----- 2. Load Example Data (Built-in Datasets) -------------------------------
# These datasets are great for learning and testing Seaborn charts.
tips = sns.load_dataset("tips")         # Tipping data from a restaurant
iris = sns.load_dataset("iris")         # Classic flower data for classification
titanic = sns.load_dataset("titanic")   # Titanic passenger information
flights = sns.load_dataset("flights")   # Flight data over years
```

```python
# ----- 3. Scatter Plot --------------------------------------------------------
# Use a scatter plot to show relationship/correlation between two continuous variables.
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", size="size", sizes=(20, 200))
plt.title("Scatter Plot: Total Bill vs Tip (Grouped by Sex/Time, Sized by Group)")  # Title for clarity
plt.xlabel("Total Bill ($)")               # Always label axes clearly
plt.ylabel("Tip ($)")
plt.show()
```

```python
# ----- 4. Line Plot -----------------------------------------------------------
# Line plots are for trends (time series, sequences, or continuous data changes).
sns.lineplot(data=flights, x="year", y="passengers")

plt.title("Line Plot: Airline Passengers Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# Example: Grouped line plot for advanced trend analysis
sns.lineplot(data=flights, x="year", y="passengers", hue="month")

plt.title("Line Plot: Passengers per Month by Year")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend(title="Month")
plt.show()
```

```python
# ----- 5. Bar Plot ------------------------------------------------------------
# Bar plots compare numeric values across categories (mean/aggregate per category).
sns.barplot(data=titanic, x="class", y="fare", hue="sex", ci=None, palette="muted")

plt.title("Bar Plot: Average Fare by Class and Gender (Titanic)")
plt.xlabel("Class")
plt.ylabel("Average Fare")
plt.show()
```

```python
# ----- 6. Histogram -----------------------------------------------------------
# Histograms show distribution of a numeric variable (frequency of value bins).
sns.histplot(data=tips, x="total_bill", bins=20, kde=True, color="teal")

plt.title("Histogram + KDE: Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
# ----- 7. Box Plot ------------------------------------------------------------
# Box plots visualize distribution, median, outliers, and compare groups/categories.
sns.boxplot(data=tips, x="day", y="total_bill", hue="time", palette="Set3")

plt.title("Box Plot: Total Bill by Day and Time")
plt.xlabel("Day of Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
# ----- 8. Violin Plot ---------------------------------------------------------
# Violin plots show distribution, density, and median like a smoothed boxplot.
sns.violinplot(x="species", y="petal_length", data=iris, palette="pastel")
plt.title("Violin Plot: Petal Length by Species (Iris)")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# ----- 9. Heatmap -------------------------------------------------------------
# Heatmaps visualize matrix-style data (correlation/aggregation tables).
corr = tips.corr(numeric_only=True)        # Compute correlation between numeric columns

sns.heatmap(corr, annot=True, cmap="YlGnBu", fmt=".2f")  # .2f = two decimals
plt.title("Heatmap: Correlation Matrix of Tips Data")
plt.show()
```

```python
# Example: Pivot table for heatmap
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
```

```python
flights_pivot
```

```python
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d')
plt.title("Flight Passengers per Month (annotated)")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
# ----- 10. Pair Plot ----------------------------------------------------------
# Pair plots explore pairwise relationships among numeric variables.
sns.pairplot(iris, hue="species", diag_kind="kde", palette="husl")
plt.suptitle("Pair Plot: Iris Dataset", y=1.02)  # Slightly raise the main title above grid
plt.show()
```

```python
# ----- 11. Joint Plot ---------------------------------------------------------
# Joint plots combine scatter plots and histograms for two variables at once.
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', hue='species', kind='scatter')

plt.suptitle("Joint Plot: Sepal Length vs Sepal Width (Iris)", y=1.02)
plt.show()
```

```python
# ----- 12. FacetGrid ----------------------------------------------------------
# FacetGrid creates small multiples ("faceted" plots) for analyzing distributions/relationships by subgroups.
g = sns.FacetGrid(tips, col="day", row="sex")     # Group by both day and sex
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip", hue="time")
g.add_legend()
plt.suptitle("FacetGrid: Total Bill vs Tip by Day/Sex/Time", y=1.05)
plt.show()
```

```python
# ----- 13. Customizing Appearance Globally ------------------------------------
# Seaborn style/themes make plots clean and modern.
sns.set_theme(style="whitegrid")  # Other options: "darkgrid", "white", "dark", "ticks"
# Change color palette
sns.set_palette("Set2")           # Use named color palettes (pastel, viridis, etc.)
```

```python
# After setting a theme/palette, your next plots adopt this style!

# ----- 14. Save Plots to File -------------------------------------------------
plt.figure(figsize=(8,6))
sns.barplot(data=iris, x="species", y="petal_length")
plt.title("Saving a Plot Example")
plt.savefig("seaborn_barplot_iris.png")    # Saves in the working directory
plt.close()                                # Closes current figure
```

```python
# =============================================================================
#                            When to Use Which Plot:
#  - Scatter Plot: Relationship/correlation (continuous variables, categories via hue/size).
#  - Line Plot: Trends over a sequence (time, continuous index).
#  - Bar Plot: Compare groups or categories (mean/aggregate metric).
#  - Histogram/KDE: Distribution of a numeric variable (frequency, density).
#  - Box/Violin Plot: Distribution + comparison across categorical groups (spread, outliers).
#  - Heatmap: Matrix data patterns (correlation, cross-tab, pivot tables).
#  - Pair Plot: Explore multiple variable relationships (EDA, clustering/classification).
#  - Joint Plot: Deep dive on two variables' relationship + their marginals.
#  - FacetGrid: Multi-panel/grouped comparisons by categorical variables.
# =============================================================================
```
